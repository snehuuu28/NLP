```python
#Cleaning the Data
```

```python
range(len(sent))
```

```python
corpus = []
for i in range(len(sent)):
    txt = re.sub('[^a-zA-Z]',' ', sent[i])
    txt = txt.lower()
    corpus.append(txt)
```

```python
corpus
```

```python
# stemming and stopwords removal
```

```python
stemmer = PorterStemmer()
```

```python
stemmer.stem('argued')
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
for i in corpus:
    words = word_tokenize(i)
```

```python
words
```

```python
for i in corpus:
    words = word_tokenize(i)
    for i in words:
        if i not in stopwords.words('english'):
            print(stemmer.stem(i))     
```

```python
#Lemmatization
```

```python
import nltk
nltk.download('wordnet')
```

```python
lemma = WordNetLemmatizer()
```

```python
for i in corpus:
    words = word_tokenize(i)
    for i in words:
        if i not in stopwords.words('english'):
            print(lemma.lemmatize(i))     
```
